```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
psnr = []
ncc = []

df1 = pd.read_csv('./Noise_Process/uniform_metrics.csv')
df2 = pd.read_csv('./Noise_Process/gaussian_metrics.csv')
```

```python
df1['结构相似度']/100
```

```python
psnr = []
ncc = []
l1= (df1['结构相似度']/100).tolist()
l2 = (df2['结构相似度']/100).tolist()
psnr.append(l1)
psnr.append(l2)

ncc.append(df1['归一化互相关'])
ncc.append(df2['归一化互相关'])
```

```python
psnr[0]
```

```python
name = ['PSNR(x10^2)','NCC']
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['font.size'] = 18
plt.figure(figsize=(12,9))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
markers = ['o', 's', '^', 'v', '>', '<']
# for i in range(len(name)):
plt.plot(df1['奇异值数量'],psnr[0],color=colors[0], marker=markers[0],label=name[0])
plt.plot(df1['奇异值数量'],ncc[0],color=colors[1], marker=markers[1],label=name[1])
# plt.axhline(y=0.55, color='r', linestyle='--', linewidth=2,label='VSM+K-means')
plt.xticks(df1['奇异值数量'])
# plt.axhline(y=0.5367, color='b', linestyle='--', linewidth=2,label='VSM+GMM')
plt.axhline(y=0.683507, color='#1f77b4', linestyle='--', linewidth=2,label='加噪图像与原始图像的PSNR(x10^2)')
plt.axhline(y=0.9853, color='#ff7f0e', linestyle='--', linewidth=2,label='加噪图像与原始图像的NCC')
plt.legend()
plt.xlabel('奇异值数量')
plt.ylabel('')
plt.title('加入均值噪声的指标变化图')
plt.savefig('./Noise_Process/uniform_metrics_curve.jpg')
```
