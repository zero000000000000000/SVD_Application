```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import os

purity = []
f = []
ari = []
name=[]

for file in os.listdir('./output'):
    if file.startswith('ch') and file.endswith('csv'):
        name.append(file)
        df = pd.read_csv('./output/'+file)
        purity.append(df['purity'])
        f.append(df['f_list'])
        ari.append(df['ari_list'])
```

```python
name
```

```python
# name=['PCA+GMM','PCA+K-means','RSVD+GMM','RSVD+K-means','SVD+GMM','SVD+K-means']
name=['PCA+K-means','RSVD+K-means','SVD+K-means']
# name=['PCA+GMM','RSVD+GMM','SVD+GMM']
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['font.size'] = 18
plt.figure(figsize=(12,9))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
markers = ['o', 's', '^', 'v', '>', '<']
for i in range(len(name)):
    plt.plot(df['k'],purity[i],color=colors[i], marker=markers[i],label=name[i])

# plt.axhline(y=0.55, color='r', linestyle='--', linewidth=2,label='VSM+K-means')
plt.xticks(df['k'])
# plt.axhline(y=0.5367, color='b', linestyle='--', linewidth=2,label='VSM+GMM')
plt.axhline(y=0.6567, color='r', linestyle='--', linewidth=2,label='VSM+K-means')
plt.legend()
plt.xlabel('奇异值数量')
plt.ylabel('纯度')
plt.title('搜狗新闻数据集的纯度比较')
plt.savefig('./output/ch_kmeans_purity')
```
