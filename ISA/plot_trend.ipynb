```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./Metrics/ch.csv')
df.head()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig,axses = plt.subplots(3,1,figsize=(10,15))
num = len(df)
col = ['purity','f_list','ari_list']
col_name = ['','F','']
# raw_metrics = [0.55,0.414,0.2946]
raw_metrics = [0.6567,0.4637,0.3581]
for i in range(3):
    axses[i].plot(df['k'],df[col[i]],'b-')
    axses[i].plot(df['k'],[raw_metrics[i]]*num,'r-')
    axses[i].set_xlabel('')
    axses[i].set_ylabel(col_name[i])
    axses[i].set_title(col_name[i]+'')

plt.tight_layout()
plt.show()
```
